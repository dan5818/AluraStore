# 4. Productos más y menos vendidos

```python
def ranking_productos(df, nombre_tienda, top=10):
    """Ranking de productos más vendidos con barras en verde (más vendido) y rojo (menos vendido)."""
    ranking = df["Producto"].value_counts().head(top)
    print(f"\nTop {top} productos más vendidos en {nombre_tienda}:")
    print(ranking)

    # Colorear: max en verde, min en rojo, resto gris
    colores = []
    for valor in ranking:
        if valor == ranking.max():
            colores.append("green")
        elif valor == ranking.min():
            colores.append("red")
        else:
            colores.append("grey")

    # 📊 Gráfica
    plt.figure(figsize=(10,5))
    bars = plt.bar(ranking.index, ranking.values, color=colores, edgecolor="black")

    # Etiquetas encima de cada barra
    for bar, valor in zip(bars, ranking.values):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
                 f"{valor}", ha="center", fontsize=9, fontweight="bold")

    plt.title(f"Top {top} productos más vendidos en {nombre_tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Número de ventas")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
    return ranking

# Ejecutar para cada tienda
ranking_t1 = ranking_productos(tienda, "Tienda 1")
ranking_t2 = ranking_productos(tienda2, "Tienda 2")
ranking_t3 = ranking_productos(tienda3, "Tienda 3")
ranking_t4 = ranking_productos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# --- 1) Función para asegurar que "Costo de envío" sea numérico ---
def asegurar_envio_numerico(df, col="Costo de envío"):
    if col not in df.columns:
        raise KeyError(f"No se encontró la columna '{col}' en el DataFrame.")
    serie = (
        df[col]
        .astype(str)
        .str.replace(r"[^\d,.\-]", "", regex=True)  # quita símbolos $ y separadores raros
        .str.replace(",", ".", regex=False)         # por si vienen comas decimales
    )
    df[col] = pd.to_numeric(serie, errors="coerce")
    return df

# Asegurar numérico en las 4 tiendas
tienda  = asegurar_envio_numerico(tienda,  "Costo de envío")
tienda2 = asegurar_envio_numerico(tienda2, "Costo de envío")
tienda3 = asegurar_envio_numerico(tienda3, "Costo de envío")
tienda4 = asegurar_envio_numerico(tienda4, "Costo de envío")

# --- 2) Cálculo del costo de envío promedio por tienda ---
prom_envio_1 = tienda["Costo de envío"].mean()
prom_envio_2 = tienda2["Costo de envío"].mean()
prom_envio_3 = tienda3["Costo de envío"].mean()
prom_envio_4 = tienda4["Costo de envío"].mean()

print(f"Costo de envío promedio - Tienda 1: ${prom_envio_1:,.2f}")
print(f"Costo de envío promedio - Tienda 2: ${prom_envio_2:,.2f}")
print(f"Costo de envío promedio - Tienda 3: ${prom_envio_3:,.2f}")
print(f"Costo de envío promedio - Tienda 4: ${prom_envio_4:,.2f}")

# DataFrame resumen
resumen_envio = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Costo de envío promedio": [prom_envio_1, prom_envio_2, prom_envio_3, prom_envio_4]
}).sort_values("Costo de envío promedio", ascending=False).reset_index(drop=True)

print("\nResumen - Costo de envío promedio por tienda:")
print(resumen_envio)

plt.figure(figsize=(10,6))

# --- 3) Gráfica: barras con colores (máximo en verde, mínimo en rojo, resto gris) ---
nombres = resumen_envio["Tienda"].tolist()
valores = resumen_envio["Costo de envío promedio"].tolist()

max_val = max(valores)
min_val = min(valores)

colores = []
for v in valores:
    if v == max_val:
        colores.append("green")
    elif v == min_val:
        colores.append("red")
    else:
        colores.append("grey")

plt.figure(figsize=(8,5))
bars = plt.bar(nombres, valores, edgecolor="black", color=colores)

# Etiquetas con el valor arriba de cada barra
for b, v in zip(bars, valores):
    plt.text(b.get_x() + b.get_width()/2, b.get_height() + (0.02 * max_val),
             f"${v:,.2f}", ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.title("Costo de envío promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Costo de envío promedio ($)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# ============================================
# Informe integral de desempeño por tienda
# (pandas + matplotlib) - Google Colab
# ============================================
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# ------------------------------
# 0) CONFIG: URLs y pesos del modelo
# ------------------------------
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Pesos del puntaje de desempeño (suma ≈ 1.0)
PESOS = {
    "ingresos": 0.40,   # más alto = mejor
    "calificacion": 0.30, # más alto = mejor
    "costo_envio": 0.20,  # más bajo = mejor
    "volumen": 0.10       # más alto = mejor (# de ventas)
}

TOP_N_PRODUCTOS = 10  # para rankings y gráficos

# ------------------------------
# 1) Utilidades
# ------------------------------
def std_cols(df):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    for col in df.select_dtypes(include="object").columns:
        df[col] = df[col].astype(str).str.strip()
    return df

def to_numeric_clean(series):
    s = (
        series.astype(str)
        .str.replace(r"[^\d,.\-]", "", regex=True) # quita símbolos
        .str.replace(",", ".", regex=False)        # coma decimal a punto
    )
    return pd.to_numeric(s, errors="coerce")

def asegurar_columnas(df):
    """Asegura y normaliza columnas clave si existen."""
    df = std_cols(df)

    # Normalizar nombres conocidos (sin cambiar los originales si ya están bien)
    # Precio
    if "Precio" in df.columns:
        df["Precio"] = to_numeric_clean(df["Precio"])

    # Costo de envío
    if "Costo de envío" in df.columns:
        df["Costo de envío"] = to_numeric_clean(df["Costo de envío"])

    # Calificación
    if "Calificación" in df.columns:
        df["Calificación"] = to_numeric_clean(df["Calificación"])

    # Fecha de Compra
    if "Fecha de Compra" in df.columns:
        df["Fecha de Compra"] = pd.to_datetime(df["Fecha de Compra"], errors="coerce", dayfirst=True)
        df["AñoMes"] = df["Fecha de Compra"].dt.to_period("M").astype(str)

    return df

def barra_extremos(idx, valores, titulo, xlabel, ylabel):
    """Bar chart destacando máximo (verde) y mínimo (rojo)."""
    max_val = np.nanmax(valores)
    min_val = np.nanmin(valores)
    colores = ["green" if v == max_val else "red" if v == min_val else "grey" for v in valores]

    plt.figure(figsize=(9,4.5))
    bars = plt.bar(idx, valores, edgecolor="black", color=colores)
    for b, v in zip(bars, valores):
        if pd.notna(v):
            plt.text(b.get_x()+b.get_width()/2, b.get_height()*1.01, f"{v:,.2f}", ha="center", fontsize=9, fontweight="bold")
    plt.title(titulo)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(axis="y", linestyle="--", alpha=0.4)
    plt.tight_layout()
    plt.show()

def ranking_productos(df, nombre_tienda, top=TOP_N_PRODUCTOS):
    """Ranking de productos más vendidos con extremos en verde/rojo."""
    if "Producto" not in df.columns:
        return None

    ranking = df["Producto"].value_counts().head(top)
    display(Markdown(f"**Top {top} productos más vendidos en {nombre_tienda}:**"))
    display(ranking)

    # colores extremo
    colores = ["green" if v == ranking.max() else "red" if v == ranking.min() else "grey" for v in ranking.values]

    plt.figure(figsize=(10,5))
    bars = plt.bar(ranking.index, ranking.values, color=colores, edgecolor="black")
    for bar, val in zip(bars, ranking.values):
        plt.text(bar.get_x()+bar.get_width()/2, val*1.01, f"{val}", ha="center", fontsize=9, fontweight="bold")
    plt.title(f"Top {top} productos más vendidos - {nombre_tienda}")
    plt.xlabel("Producto")
    plt.ylabel("Número de ventas")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
    return ranking

# ------------------------------
# 2) Carga
# ------------------------------
t1 = asegurar_columnas(pd.read_csv(url1))
t2 = asegurar_columnas(pd.read_csv(url2))
t3 = asegurar_columnas(pd.read_csv(url3))
t4 = asegurar_columnas(pd.read_csv(url4))

t1["Tienda"] = "Tienda 1"
t2["Tienda"] = "Tienda 2"
t3["Tienda"] = "Tienda 3"
t4["Tienda"] = "Tienda 4"

tiendas = [t1, t2, t3, t4]
nombres = [ "Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4" ]

# ------------------------------
# 3) Indicadores por tienda
# ------------------------------
# Ingresos, volumen, calificación promedio, costo envío promedio
ingresos = []
volumen = []
calif = []
envio = []

for df in tiendas:
    ingresos.append(df["Precio"].sum() if "Precio" in df.columns else np.nan)
    volumen.append(len(df))
    calif.append(df["Calificación"].mean() if "Calificación" in df.columns else np.nan)
    envio.append(df["Costo de envío"].mean() if "Costo de envío" in df.columns else np.nan)

resumen = pd.DataFrame({
    "Tienda": nombres,
    "Ingreso Total": ingresos,
    "Ventas (conteo)": volumen,
    "Calificación Promedio": calif,
    "Costo de Envío Promedio": envio
})
display(Markdown("### Resumen de indicadores por tienda"))
display(resumen)

# ------------------------------
# 4) Gráficas globales
# ------------------------------
# 4.1 Ingresos totales por tienda
barra_extremos(
    resumen["Tienda"],
    resumen["Ingreso Total"].values,
    "Ingresos totales por tienda",
    "Tienda", "Ingresos ($)"
)

# 4.2 Calificación promedio por tienda
plt.figure(figsize=(9,4.5))
bars = plt.bar(resumen["Tienda"], resumen["Calificación Promedio"], edgecolor="black")
for b, v in zip(bars, resumen["Calificación Promedio"]):
    if pd.notna(v):
        plt.text(b.get_x()+b.get_width()/2, v*1.01, f"{v:.2f}", ha="center", fontsize=9, fontweight="bold")
plt.title("Calificación promedio por tienda")
plt.xlabel("Tienda")
plt.ylabel("Calificación")
plt.ylim(0, 5)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# 4.3 Dispersión de costos de envío (cada tienda color distinto)
plt.figure(figsize=(10,6))
if "Costo de envío" in t1.columns: plt.scatter(range(len(t1)), t1["Costo de envío"], alpha=0.6, label="Tienda 1")
if "Costo de envío" in t2.columns: plt.scatter(range(len(t2)), t2["Costo de envío"], alpha=0.6, label="Tienda 2")
if "Costo de envío" in t3.columns: plt.scatter(range(len(t3)), t3["Costo de envío"], alpha=0.6, label="Tienda 3")
if "Costo de envío" in t4.columns: plt.scatter(range(len(t4)), t4["Costo de envío"], alpha=0.6, label="Tienda 4")
plt.title("Dispersión de costos de envío por tienda")
plt.xlabel("Índice de venta")
plt.ylabel("Costo de envío ($)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# 4.4 Boxplot costo de envío por tienda
data_envio = []
labels_envio = []
for df, name in zip(tiendas, nombres):
    if "Costo de envío" in df.columns:
        data_envio.append(df["Costo de envío"].dropna())
        labels_envio.append(name)
if data_envio:
    plt.figure(figsize=(10,6))
    plt.boxplot(data_envio, labels=labels_envio, showmeans=True, notch=True)
    plt.title("Distribución de costos de envío por tienda (Boxplot)")
    plt.ylabel("Costo de envío ($)")
    plt.grid(axis="y", linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()

# ------------------------------
# 5) Categorías y productos (por tienda)
# ------------------------------
col_cat = "Categoría del Producto" if "Categoría del Producto" in t1.columns else None
for df, name in zip(tiendas, nombres):
    display(Markdown(f"### {name} — Ventas por categoría"))
    if col_cat and col_cat in df.columns:
        ventas_cat = df[col_cat].value_counts()
        display(ventas_cat)

        # gráfico barras categorías
        plt.figure(figsize=(10,5))
        bars = plt.bar(ventas_cat.index, ventas_cat.values, edgecolor="black")
        for b, v in zip(bars, ventas_cat.values):
            plt.text(b.get_x()+b.get_width()/2, v*1.01, f"{int(v)}", ha="center", fontsize=9, fontweight="bold")
        plt.title(f"Ventas por categoría - {name}")
        plt.xlabel("Categoría")
        plt.ylabel("Número de ventas")
        plt.xticks(rotation=45, ha="right")
        plt.tight_layout()
        plt.show()
    else:
        display(Markdown("_No se encontró la columna de categoría._"))

    # Ranking de productos con colores
    ranking_productos(df, name, top=TOP_N_PRODUCTOS)

# ------------------------------
# 6) Puntaje de desempeño y recomendación
# ------------------------------
# Normalización min-max por métrica
def minmax(series):
    s = series.astype(float)
    mx, mn = s.max(), s.min()
    if pd.isna(mx) or pd.isna(mn) or mx == mn:
        # si no hay variación, devuelvo 0.5 para no sesgar
        return pd.Series([0.5]*len(s), index=s.index)
    return (s - mn) / (mx - mn)

sc_ing = minmax(resumen["Ingreso Total"])                   # más es mejor
sc_cal = minmax(resumen["Calificación Promedio"])           # más es mejor
sc_vol = minmax(resumen["Ventas (conteo)"])                 # más es mejor
# costo de envío invertido (menor mejor)
sc_env = 1 - minmax(resumen["Costo de Envío Promedio"])     # menos es mejor

score = (
    PESOS["ingresos"]     * sc_ing +
    PESOS["calificacion"] * sc_cal +
    PESOS["costo_envio"]  * sc_env +
    PESOS["volumen"]      * sc_vol
)

resumen["Score Desempeño (0-1)"] = score.round(3)
display(Markdown("### Puntaje de desempeño (0=peor, 1=mejor)"))
display(resumen[["Tienda","Ingreso Total","Ventas (conteo)","Calificación Promedio","Costo de Envío Promedio","Score Desempeño (0-1)"]])

tienda_mejor = resumen.loc[resumen["Score Desempeño (0-1)"].idxmax(), "Tienda"]
tienda_peor  = resumen.loc[resumen["Score Desempeño (0-1)"].idxmin(), "Tienda"]

display(Markdown(f"> **Conclusión rápida**: Mejor desempeño: **{tienda_mejor}** · Menor desempeño (recomendada para vender): **{tienda_peor}**"))

# ------------------------------
# 7) Construcción de Informe Final (Markdown)
# ------------------------------
def fmt_money(x):
    return f"${x:,.2f}" if pd.notna(x) else "N/D"

def top_y_bottom_cat(df):
    if "Categoría del Producto" not in df.columns:
        return ("N/D","N/D")
    vc = df["Categoría del Producto"].value_counts()
    if vc.empty:
        return ("N/D","N/D")
    return (vc.idxmax(), vc.idxmin())

def top_y_bottom_prod(df):
    if "Producto" not in df.columns:
        return ("N/D","N/D")
    vc = df["Producto"].value_counts()
    if vc.empty:
        return ("N/D","N/D")
    return (vc.idxmax(), vc.idxmin())

bloques = []
intro = (
    "# Informe Final de Desempeño por Tienda\n\n"
    "## Introducción\n"
    "Este informe evalúa el desempeño de las **cuatro tiendas** del Sr. Juan con base en: "
    "**ingresos totales**, **categorías más/menos vendidas**, **calificación promedio de clientes**, "
    "**productos más/menos vendidos** y **costo de envío promedio**. "
    "El objetivo es identificar **qué tienda tiene el menor desempeño** y, por tanto, "
    "**cuál sería la candidata a vender**, justificando la recomendación con datos y visualizaciones.\n\n"
    "## Metodología\n"
    "1. Se integraron y limpiaron los datos (precios, calificaciones y costos de envío a formato numérico; fechas con `dayfirst=True`).\n"
    "2. Se calcularon métricas por tienda y se generaron visualizaciones (barras, dispersión y boxplots).\n"
    "3. Se construyó un **puntaje de desempeño** (0–1) mediante normalización min–max con pesos:\n"
    f"   - Ingresos: {PESOS['ingresos']:.2f} · Calificación: {PESOS['calificacion']:.2f} · "
    f"Costo de envío (menor mejor): {PESOS['costo_envio']:.2f} · Volumen: {PESOS['volumen']:.2f}.\n"
    "4. La tienda con **menor puntaje** se considera la de **peor desempeño** (candidata a vender).\n"
)

bloques.append(intro)

desarrollo = "## Resultados y hallazgos por tienda\n"
for df, name, ing, vol, cal, env in zip(tiendas, nombres, ingresos, volumen, calif, envio):
    cat_top, cat_bot = top_y_bottom_cat(df)
    prod_top, prod_bot = top_y_bottom_prod(df)
    desarrollo += (
        f"\n### {name}\n"
        f"- **Ingresos totales**: {fmt_money(ing)}\n"
        f"- **Volumen de ventas (conteo)**: {vol}\n"
        f"- **Calificación promedio**: {cal:.2f} \n" if pd.notna(cal) else f"- **Calificación promedio**: N/D\n"
    )
    desarrollo += f"- **Costo de envío promedio**: {fmt_money(env)}\n"
    desarrollo += f"- **Categoría más vendida**: {cat_top}\n"
    desarrollo += f"- **Categoría menos vendida**: {cat_bot}\n"
    desarrollo += f"- **Producto más vendido**: {prod_top}\n"
    desarrollo += f"- **Producto menos vendido**: {prod_bot}\n"

bloques.append(desarrollo)

tabla_scores = "## Comparativa y puntaje de desempeño\n"
tabla_scores += resumen[["Tienda","Ingreso Total","Ventas (conteo)","Calificación Promedio","Costo de Envío Promedio","Score Desempeño (0-1)"]].to_markdown(index=False)
bloques.append(tabla_scores)

conclusion = (
    "\n\n## Conclusión y recomendación\n"
    f"Con base en el **puntaje de desempeño** y los indicadores evaluados, "
    f"la tienda con **mejor** desempeño es **{tienda_mejor}**, mientras que la tienda con **menor** desempeño es **{tienda_peor}**.\n\n"
    f"**Recomendación:** Considerar **vender {tienda_peor}**. Esta decisión se justifica por:\n"
    "- Ingresos relativamente menores frente a las otras tiendas.\n"
    "- Calificación promedio que no compensa el rezago en ingresos/volumen.\n"
    "- Costo de envío promedio que no presenta ventaja comparativa suficiente (o incluso es superior).\n"
    "- Mix de categorías y productos menos favorable (según los rankings y su concentración).\n\n"
    "Si la prioridad estratégica del Sr. Juan fuera **experiencia del cliente** por sobre ingresos, puede ajustarse el modelo incrementando el peso de la **calificación** "
    "y reduciendo el de **ingresos**. En ese caso, el ranking podría cambiar. No obstante, con los pesos actuales, "
    f"**{tienda_peor}** es la candidata más razonable para vender.\n"
)
bloques.append(conclusion)

informe_md = "\n\n".join(bloques)

# Mostrar y guardar informe
display(Markdown(informe_md))
with open("informe_tiendas.md", "w", encoding="utf-8") as f:
    f.write(informe_md)

print("\nArchivo guardado: informe_tiendas.md (Markdown)")
```
